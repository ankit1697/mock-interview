```python
import torchaudio as ta
from chatterbox.tts import ChatterboxTTS

# Load the pretrained Chatterbox model; use "cuda" if available, else "cpu"
tts = ChatterboxTTS.from_pretrained(device="mps")

# test set vs validation set in details
text = "The test set contains examples that the model has never seen during training or validation, allowing for an unbiased evaluation of its performance in real-world scenarios."
text+= " The validation set is used during training to tune hyperparameters and make decisions about model architecture, helping to prevent overfitting and improve generalization."
text+= " Both sets are crucial for assessing the model's ability to generalize to new, unseen data."

# Generate speech waveform with your voice cloned from a reference audio file "reference.wav"
wav = tts.generate(text, audio_prompt_path="reference.wav")

# Save the resulting wave file
ta.save("cloned.wav", wav, tts.sr)
```
